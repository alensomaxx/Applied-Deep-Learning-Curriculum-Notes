Data Preprocessing

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('/content/Churn_Modelling.csv')
dataset.tail()
```

```python
print(dataset)
```

```python
#Split Dataset into X and y
X = pd.DataFrame(dataset.iloc[:, 3:13].values) #Independent variables
y = dataset.iloc[:, 13].values #dependent variables
```

```python
X.head(10)
```

```python
print(X)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_2 = LabelEncoder()
X.loc[:, 2] = labelencoder_X_2.fit_transform(X.iloc[:, 2])
```

```python
print(X)
```

```python
labelencoder_X_1 = LabelEncoder()
X.loc[:, 1] = labelencoder_X_1.fit_transform(X.iloc[:, 1])
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder #scikit-learn
from sklearn.compose import ColumnTransformer
```

```python
#Country column
ct = ColumnTransformer([("Country", OneHotEncoder(), [1])], remainder = 'passthrough')
X = ct.fit_transform(X)

# Male/Female
labelencoder_X = LabelEncoder()
X[:, 2] = labelencoder_X.fit_transform(X[:, 2])
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

### **Build Artificial Neural Network**

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 12))
classifier.add(Dense(6, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
