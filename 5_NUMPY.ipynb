```python
import numpy as np

arr = np.array ([[1, 2, 3],
                 [4, 5, 6]])

print(arr)

#type of array

print("Array of the type: ", type(arr))

# dimensions

print("No. of dimensions: ", arr.ndim)

# Shape

print("Shape of array: ", arr.shape)

# Size

print("Size of array: ", arr.size)
```
